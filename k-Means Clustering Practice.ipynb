```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers = 4, cluster_std=1.8,
                 random_state =101)
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap="rainbow")
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

if we were working with real data, we would not know the real lables, only the predicted lables. But since this is generated data we can figure out the accuracy of this test by comparing it to the real lables we generated.

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey = True, figsize=(10,6))
ax1.set_title("K Means")
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap="rainbow")

ax2.set_title("Original")
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap="rainbow")
```
